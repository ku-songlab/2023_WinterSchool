```python
#계단 함수 정의하기

def step_function(x):
  if x >=0:
    return 1
  else:
    return 0

#계단 함수 값 입력하기

step_function()
```

```python
#필요한 library 불러오기

import numpy as np
import matplotlib.pyplot as plt

#relu 함수 정의하기

def relu(x):
    return np.maximum(0, x)     


x = np.arange(-10, 10, 0.1)
y = relu(x)

#시각화

plt.plot(x, y)
plt.show()
```

```python
#필요한 library 불러오기

import numpy as np
import matplotlib.pyplot as plt

#sigmoid 함수 정의하기

def sigmoid(x):
    return 1 / (1 +np.exp(-x))
 
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

#시각화

plt.plot(x, y)
plt.show()
```
